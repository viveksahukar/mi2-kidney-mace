```python
class ECGDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data
        # Convert labels to Python scalars if they are tensors
        if isinstance(labels[0], torch.Tensor):
            self.labels = [label.item() for label in labels]
        else:
            self.labels = labels
        self.targets = self.labels  # For ImbalancedDatasetSampler

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # Check if the data is already a tensor
        if isinstance(self.data[idx], torch.Tensor):
            data_tensor = self.data[idx].type(torch.float32)
        else:
            data_tensor = torch.tensor(self.data[idx], dtype=torch.float32)
        # Ensure labels are a torch.Tensor
        # Convert to long if they are classification labels
        label_tensor = torch.tensor(self.labels[idx], dtype=torch.float32)
        return data_tensor, label_tensor
```

```python
class ECGCNNAttn(nn.Module):
    def __init__(self, num_channels=12, signal_length=6, num_heads=2):
        super(ECGCNNAttn, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=num_channels, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm1d(num_features=32)
        self.dropout1 = nn.Dropout(0.5)
        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm1d(num_features=64)
        self.dropout2 = nn.Dropout(0.5)
        
        # Assuming that the output of conv2 has a size of 'signal_length // 4'
        # Adjust the size accordingly if your architecture changes
        self.attention = nn.MultiheadAttention(embed_dim=64, num_heads=num_heads)
        
        # Flatten layer will be handled in the forward method
        # The input features to fc1 should be adjusted according to the output of the attention layer
        self.fc1 = nn.Linear(64 * (signal_length // 4), 64)
        self.bn_fc1 = nn.BatchNorm1d(64)
        self.dropout_fc1 = nn.Dropout(0.8)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        # Convolutional layers with pooling and batch norm
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.dropout1(x)
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.dropout2(x)
        
        # Permute the dimensions to fit the input requirements of nn.MultiHeadAttention
        # From (batch_size, num_channels, signal_length) to (signal_length, batch_size, num_channels)
        x = x.permute(2, 0, 1)
        
        # Apply the attention module
        attn_output, _ = self.attention(x, x, x)
        
        # Permute back to (batch_size, signal_length, num_channels) to flatten
        x = attn_output.permute(1, 2, 0).flatten(start_dim=1)
        
        # Fully connected layers with batch norm and dropout
        x = F.relu(self.bn_fc1(self.fc1(x)))
        x = self.dropout_fc1(x)
        x = self.fc2(x)

        return x
```

```python
class ECGCNNAttnFirst(nn.Module):
    def __init__(self, num_channels=12, signal_length=6, num_heads=2):
        super(ECGCNNAttnFirst, self).__init__()

        # Attention layer first
        # The input embedding dimension for MultiHeadAttention must match the number of channels
        self.attention = nn.MultiheadAttention(embed_dim=num_channels, num_heads=num_heads)
        
        # Convolutional layers
        self.conv1 = nn.Conv1d(in_channels=num_channels, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm1d(num_features=32)  # Batch normalization
        self.pool = nn.MaxPool1d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm1d(num_features=64)  # Batch normalization
        self.dropout = nn.Dropout(0.5)  # Dropout after convolutional layers

        # Fully connected layers after convolutional layers
        self.fc1 = nn.Linear(64 * (signal_length // 4), 64)  # Adjusted for pooling layers
        self.bn_fc1 = nn.BatchNorm1d(64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        # Apply attention first
        # Need to permute x to match the input shape for attention (seq_len, batch, features)
        x = x.permute(2, 0, 1)
        attn_output, _ = self.attention(x, x, x)
        # Permute back to (batch, features, seq_len)
        x = attn_output.permute(1, 2, 0)

        # Then apply convolutional layers
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))

        # Flatten the output for the dense layers
        x = x.view(x.size(0), -1)
        x = self.dropout(x)

        # Apply dense layers
        x = F.relu(self.bn_fc1(self.fc1(x)))
        x = self.fc2(x)

        return x
```

```python
class HeartNetModel(nn.Module):
    def __init__(self, in_channels):
        super(HeartNetModel, self).__init__()
        # Convolutional block
        self.conv1 = nn.Conv1d(in_channels, 64, kernel_size=7, stride=1, padding='same')
        self.bn1 = nn.BatchNorm1d(64)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.5)

        self.conv2 = nn.Conv1d(64, 128, kernel_size=5, stride=1, padding='same')
        self.bn2 = nn.BatchNorm1d(128)
        self.relu2 = nn.ReLU()

        self.conv3 = nn.Conv1d(128, 256, kernel_size=3, stride=1, padding='same')
        self.bn3 = nn.BatchNorm1d(256)
        self.softmax3 = nn.Softmax(dim=1)
        self.dropout3 = nn.Dropout(0.5)

        self.conv4 = nn.Conv1d(256, 128, kernel_size=1, stride=1, padding='same')

        # Since the model specifies multi-head attention and a global average pooling,
        # we need to flatten the output from conv layers before applying these.
        # self.flatten = nn.Flatten()

        # Multi-Head Attention Block
        # PyTorch doesn't have a direct multi-head attention for 1d convolutions.
        # A workaround is to use the nn.MultiheadAttention which works on sequences.
        # We would need to adjust dimensions appropriately.
        self.multihead_attn = nn.MultiheadAttention(embed_dim=128, num_heads=2, dropout=0.5)

        # Output Block
        self.global_avg_pooling = nn.AdaptiveAvgPool1d(1)
        self.fc = nn.Linear(128, 1)

    def forward(self, x):
        # Convolutional block
        x = self.conv1(x)
        x = self.bn1(x)
        x = self.relu1(x)
        x = self.dropout1(x)

        x = self.conv2(x)
        x = self.bn2(x)
        x = self.relu2(x)

        x = self.conv3(x)
        x = self.bn3(x)
        x = self.softmax3(x)
        x = self.dropout3(x)

        x = self.conv4(x)
        x = x.permute(2, 0, 1)  # Adjust dimensions for multihead attention (L, N, E)
        attn_output, _ = self.multihead_attn(x, x, x)
        x = attn_output.permute(1, 2, 0)  # Back to (N, E, L)

        # Global average pooling and classifier
        x = self.global_avg_pooling(x)
        x = x.view(x.size(0), -1)  # Flatten the output
        x = self.fc(x)
        return x

# Assuming the input data dimensions are [batch_size, no_of_leads, length]
# For 12-lead ECG, the number of in_channels will be 12
# For binary classification, the number of classes will be 2


```

```python
class ECGTransformer(nn.Module):
    def __init__(self, num_channels=12, signal_length=6, num_heads=2, num_encoder_layers=2):
        super(ECGTransformer, self).__init__()
        # Transformer Encoder
        encoder_layer = nn.TransformerEncoderLayer(d_model=num_channels, nhead=num_heads)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_encoder_layers)
        
        # Positional Encoding (optional, depending on your use case)
        # ...

        # Final Dense layers for prediction
        self.fc1 = nn.Linear(num_channels * signal_length, 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        # Apply positional encoding if necessary
        # ...

        # Permute x to shape (signal_length, batch_size, num_channels) for Transformer
        x = x.permute(2, 0, 1)

        # Apply Transformer Encoder
        x = self.transformer_encoder(x)

        # Flatten and apply Dense layers
        x = x.permute(1, 2, 0).flatten(start_dim=1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)

        return x
```

```python
# Hyperparameters
num_epochs = 50
learning_rate = 1e-5

# Model, criterion, and optimizer
# model = ECGCNNAttnFirst().to(device)
# model = HeartNetModel(in_channels=12).to(device)
# model = ECGCNNAttn().to(device)
model = ECGTransformer().to(device)
criterion = nn.BCEWithLogitsLoss()  # Assuming pos_weight is already defined
n = 2 
pos_weight = pos_weight / n
optimizer = optim.SGD(model.parameters(pos_weight), lr=learning_rate)

# Scheduler for cyclic learning rate
scheduler = CyclicLR(optimizer, base_lr=1e-5, max_lr=1e-3, step_size_up=5, 
                     step_size_down=20, mode="triangular2", cycle_momentum=True)


# For storing loss values
train_losses = []
val_losses = []

for epoch in range(num_epochs):
    model.train()  # Set model to training mode
    running_loss = 0.0

    for i, (inputs, labels) in enumerate(train_loader):
        # Move inputs and labels to the device the model is on
        inputs, labels = inputs.to(device), labels.to(device)
        

        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, labels.unsqueeze(1).float())  # Adjust label shape and type if necessary

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        scheduler.step()

        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)
    train_losses.append(train_loss)

    # Validation loop
    model.eval()  # Set model to evaluation mode
    running_loss = 0.0
    
    # Collect all labels and model outputs
    true_labels = []
    predictions = []

    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels.unsqueeze(1).float())  # Adjust label shape and type if necessary
            running_loss += loss.item()

            # Apply sigmoid since BCEWithLogitsLoss combines a Sigmoid layer
            predicted_probs = torch.sigmoid(outputs).cpu().numpy()
            predictions.extend(predicted_probs.flatten().tolist())
            true_labels.extend(labels.cpu().numpy().flatten().tolist())

    val_loss = running_loss / len(val_loader)
    val_losses.append(val_loss)

    print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Validation Loss: {val_loss:.4f}")

# After training, you can plot the training and validation loss curves
```
